```python
from sklearn.datasets import make_blobs
import matplotlib.pyplot as plt 
import pandas as pd
```

```python
X, y = make_blobs(n_samples=150, centers=3, random_state=0, cluster_std=.5) # shrinking the cluster_std, the standard deviation, shrinks the cluster size because the average distance from the mean gets smaller or larger
```

```python
X = pd.DataFrame(X)
y = pd.DataFrame(y)
```

```python
plt.scatter(
    x=X[0],
    y=X[1],
    c=y
)
```
